# Research Question

-  Is there a relationship between average monthly precipitation levels and average monthly commercial aircraft fuel consumption in the last 10 years in the United States?
-  Is there a relationship between average flight fuel consumption and average ticket price for commercial aircraft in the last 10 years in the United States?



## Background and Prior Work

In a globalizing world, air travel has become increasingly important not only for economic purposes, but for individual purposes as well. However, since the early 2000s, fuel prices have continued to increase, making it ever more important to optimize fuel efficiency. Consumers are increasingly concerned about their carbon footprint when they fly, and some airlines are offering dubious, albeit optional, carbon offset add-ons to their tickets.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Not only do CO2 emissions from commercial flights contribute to air pollution, but ticket costs are also affected by fuel prices and consumption rates.


According to an article by the firm McKinsey & Company, between 2005 and 2019 airlines managed to decrease fuel per passenger-kilometer by 39%.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Roughly half of this decrease owes itself to an increase in seat density and percentage of seats per flight sold. Airlines increased the number of passengers per plane by 82% in 2005 and by 88% in 2019.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Next, fleet upgrades account for a little less than half of this decrease in fuel per passenger-kilometer.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) As aircrafts have become more aerodynamic, they’ve become more fuel efficient as well. Other factors like route optimization accounts for less than 7% of the improved fuel mileage. As technological advancements slow down, airlines inevitably have begun to hit a plateau in improving fuel efficiency. Likewise, increasing aircraft density can only go so far.

Similarly, the paper *Fuel consumption optimization in air transport* by Singh and Sharma examines historical trends of fuel consumption optimization (FCO). They identify four factors that affect fuel consumption: aircraft technology & design, aviation operations & infrastructure, socioeconomic & policy measures, and alternate fuels & fuel properties.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) The authors found that “Fuel consumption is one of major direct operating cost parameters in the air transport industry” and accounts for roughly 28% of operating costs, but in the future it may account for as much as 45%.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) Similar to the findings of McKinsey, while jet fuel has increased significantly since 2005, fuel efficiency improvements have slowed down since the 1980s. Airline route optimization is incredibly complex, bringing together diverse factors such as aircraft performance, weather, route and regulatory restrictions, luggage and passenger capacity, and global fuel and travel economics.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) For the scope of our project, we will be focusing mainly on the relationship between cabin density and fuel consumption per passenger-kilometer. We hope to understand how external factors such as weather also play into fuel consumption and subsequently ticket prices. We hope to study precipitation levels in origin and destination airports, fuel consumption, and ticket prices of those flights. Our goal is to determine whether future aircraft manufacturing efforts should aim to make airplanes more adaptable to poor weather conditions by determining if weather is a factor affecting fuel consumption.

### Question Part 1: Precipitation & Fuel Consumption

The first part of our research question asks whether there is a relationship between average precipitation in the U.S. and average fuel consumption in the last 10 years, to see how this type of weather affects fuel efficiency.

```python

# Convert 'Year_Month' in dataset3 to datetime format
dataset3['Year_Month'] = pd.to_datetime(dataset3['Year_Month'], format='%Y-%m')

# Now, let's assume we want to merge dataset1 and dataset3 based on 'Year_Month'
# Note: This step assumes both datasets have a 'Year_Month' column in datetime format
merged_dataset = pd.merge(dataset1, dataset3, on='Year_Month', how='inner')

correlation_matrix = merged_dataset[['Consumption(million gallons)', 'Value']].corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation between Precipitation and Fuel Consumption')
plt.show()
```

Based on the heatmap provided, the correlation coefficient between precipitation (Value) and fuel consumption (Consumption in million gallons) is approximately 0.23.
A correlation coefficient of 0.23 indicates a weak positive linear relationship between precipitation and fuel consumption. This suggests that there might be a slight tendency for fuel consumption to increase as precipitation increases, but the relationship is not strong. With such a weak correlation, we can say that precipitation alone is not a strong predictor of fuel consumption. It is likely that other factors have a more significant impact on fuel consumption than precipitation does. Future research initiatives should focus on correlation of weather and fuel consumption in more extreme weather conditions since normal levels of rain do not have a significant impact on fuel consumption. However, in extreme weather cases, flights often will be delayed for safety reasons. A potential further area of research could then be on ticket price and fuel consumption fluctuation during the El nino years in particular. From the data we have seen so far, however, rain is not a strong contributor to fuel usage.

```python
plt.figure(figsize=(12, 8))
sns.regplot(x='Value', y='Consumption(million gallons)', data=merged_dataset)
plt.title('Scatter Plot with Regression Line: Precipitation vs. Fuel Consumption')
plt.xlabel('Average Precipitation (inches)')
plt.ylabel('Fuel Consumption (million gallons)')
plt.show()
```

The scatter plot above illustrates the relationship between average precipitation (in inches) and fuel consumption (in million gallons). By incorporating a regression line, we aim to quantify the linear relationship between these two variables. While the plot indicates a certain level of positive association—meaning that fuel consumption seems to increase with rising precipitation levels—the dispersion of data points also suggests that this relationship is not very tight. This further supports our earlier conclusion based on the correlation coefficient analysis, indicating a weak positive relationship between precipitation and fuel consumption. However, precipitation alone is not a strong predictor of fuel consumption. This might imply that other, more significant factors are at play in influencing fuel consumption.

### Question Part 2: Fuel Consumption & Ticket Prices

The second part of our research question asks whether there is a relationship between fuel consumption and ticket prices in the U.S. over the last 10 years.

```python
# Summarizing key trends for Dataset 1: Fuel Consumption and Cost
fuel_consumption_summary = dataset1.groupby('Year').agg({
    'Consumption(million gallons)': 'sum',
    'Cost(million dollars)': 'sum',
    'Cost per Gallon (dollars)': 'mean'
}).reset_index()

# Correcting column names for Dataset 2 and summarizing ticket prices
ticket_prices_summary = dataset2.groupby('Year').agg({
    'U.S. Average (Current $) ': 'mean',  # Corrected column name
    'U.S. Average (Inflation-Adjusted $) ': 'mean'  # Corrected column name
}).reset_index()

common_years = set(fuel_consumption_summary['Year']).intersection(set(ticket_prices_summary['Year']))
fuel_summary_aligned = fuel_consumption_summary[fuel_consumption_summary['Year'].isin(common_years)]
ticket_prices_aligned = ticket_prices_summary[ticket_prices_summary['Year'].isin(common_years)]

comparison_data = pd.merge(fuel_summary_aligned, ticket_prices_aligned, on='Year', how='inner')

# Plotting
fig, ax1 = plt.subplots(figsize=(14, 8))

color = 'tab:red'
ax1.set_xlabel('Year')
ax1.set_ylabel('Cost per Gallon (dollars)', color=color)
ax1.plot(comparison_data['Year'], comparison_data['Cost per Gallon (dollars)'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('U.S. Average (Inflation-Adjusted $)', color=color)  # we already handled the x-label with ax1
ax2.plot(comparison_data['Year'], comparison_data['U.S. Average (Inflation-Adjusted $) '], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Comparison of Fuel Cost per Gallon and U.S. Average Ticket Price (Inflation-Adjusted $) Over Time')
plt.show()
```

Initially, there appears to be an inverse relationship, as the cost per gallon of fuel (shown on the left y-axis) decreases and the average ticket price (shown on the right y-axis) increases. This could suggest that other factors are having a more significant impact on ticket prices than fuel costs alone during this period. Around 2020, there is a noticeable sharp decline in both fuel cost and ticket price, which then recover sharply. This synchronized movement post-2020 suggests a positive relationship during this period, likely influenced by external events (e.g., the COVID-19 pandemic) that simultaneously impacted fuel prices and ticket prices. After the 2020 decline, the rebound in both fuel costs and ticket prices indicates a positive relationship, where increases in fuel costs are accompanied by increases in ticket prices. This suggests that during the recovery phase, fuel costs may have had a more direct impact on ticket prices. Considering the entire timeline, the relationship between fuel costs and ticket prices does not appear to be consistently positive. While the post-2020 recovery phase shows a positive relationship, the pre-2020 trend suggests that ticket prices may be influenced by a complex set of factors, of which fuel is only one part. In particular, during 2022 there was a sharp increase in cost per gallon prices in 2022 in both fuel consumption and ticket prices, but the rate of increase was smaller in ticket prices. This suggests that ticket prices do not increase at the same rate that cost per gallon does.
