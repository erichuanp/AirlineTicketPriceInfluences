# Research Question

-  Is there a relationship between average monthly precipitation levels and average monthly commercial aircraft fuel consumption in the last 10 years in the United States?
-  Is there a relationship between average flight fuel consumption and average ticket price for commercial aircraft in the last 10 years in the United States?



## Background and Prior Work

In a globalizing world, air travel has become increasingly important not only for economic purposes, but for individual purposes as well. However, since the early 2000s, fuel prices have continued to increase, making it ever more important to optimize fuel efficiency. Consumers are increasingly concerned about their carbon footprint when they fly, and some airlines are offering dubious, albeit optional, carbon offset add-ons to their tickets.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Not only do CO2 emissions from commercial flights contribute to air pollution, but ticket costs are also affected by fuel prices and consumption rates.


According to an article by the firm McKinsey & Company, between 2005 and 2019 airlines managed to decrease fuel per passenger-kilometer by 39%.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Roughly half of this decrease owes itself to an increase in seat density and percentage of seats per flight sold. Airlines increased the number of passengers per plane by 82% in 2005 and by 88% in 2019.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Next, fleet upgrades account for a little less than half of this decrease in fuel per passenger-kilometer.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) As aircrafts have become more aerodynamic, they’ve become more fuel efficient as well. Other factors like route optimization accounts for less than 7% of the improved fuel mileage. As technological advancements slow down, airlines inevitably have begun to hit a plateau in improving fuel efficiency. Likewise, increasing aircraft density can only go so far.

Similarly, the paper *Fuel consumption optimization in air transport* by Singh and Sharma examines historical trends of fuel consumption optimization (FCO). They identify four factors that affect fuel consumption: aircraft technology & design, aviation operations & infrastructure, socioeconomic & policy measures, and alternate fuels & fuel properties.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) The authors found that “Fuel consumption is one of major direct operating cost parameters in the air transport industry” and accounts for roughly 28% of operating costs, but in the future it may account for as much as 45%.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) Similar to the findings of McKinsey, while jet fuel has increased significantly since 2005, fuel efficiency improvements have slowed down since the 1980s. Airline route optimization is incredibly complex, bringing together diverse factors such as aircraft performance, weather, route and regulatory restrictions, luggage and passenger capacity, and global fuel and travel economics.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) For the scope of our project, we will be focusing mainly on the relationship between cabin density and fuel consumption per passenger-kilometer. We hope to understand how external factors such as weather also play into fuel consumption and subsequently ticket prices. We hope to study precipitation levels in origin and destination airports, fuel consumption, and ticket prices of those flights. Our goal is to determine whether future aircraft manufacturing efforts should aim to make airplanes more adaptable to poor weather conditions by determining if weather is a factor affecting fuel consumption.

# Hypothesis


Experience suggests that very increased precipitation leads to more frequent encounters with unstable air currents for aircraft, necessitating changes in flight altitude and routes, but small amounts of precipitation probably have little impact on flight routes. This likely results in underestimations of optimal fuel consumption and higher actual fuel usage in extreme weather. To offset the additional operational costs, airlines are expected to raise ticket prices when weather is less predictable (such as in the winter) and fuel consumption is high.

Using this data, we can examine the relationship between average fuel consumption and average precipitation in the U.S. on any given month. The year and month data are objects/strings while the consumption and cost columns are floats.

Dataset #2: This dataset describes the average airline fare of domestic flights in the U.S. each quarter from 1995 to 2023. It also includes an inflation-adjusted price amount. Because we couldn't find monthly fare data, we may need to compare by quarter with average fuel consumption during the corresponding months. This will require combining (averaging) the months that correspond with each quarter in the other two datasets to make a comparable time range for our analysis. The year and quarter data are objects/strings while the price columns are floats.

Dataset #3: This dataset gives us average precipitation across the contiguous U.S. in inches by month. Data is available beginning in 1895, but for our analysis we will only look at the past 10 years. The year and month data are objects/strings while the precipitation and anomaly columns are floats.

## Dataset #1: Airline Fuel Cost and Consumption

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION 
dataset1 = pd.read_csv('Datasets/dataset1.csv')
dataset1 = dataset1[(dataset1['Year'] >= '2014') & (dataset1['Year'] <= '2023')]  # limit to ten years 2014-2023
dataset1 = dataset1[~((dataset1['Month'].str.lower() == 'total') | (dataset1['Year'].str.contains('% Chg over')))]  # delete rows that are not monthly
month_to_num = {
    'January': '1', 'February': '2', 'March': '3', 'April': '4', 'May': '5', 'June': '6',
    'July': '7', 'August': '8', 'September': '9', 'October': '10', 'November': '11', 'December': '12'
}
dataset1['Month'] = dataset1['Month'].str.strip().apply(lambda x: month_to_num.get(x, None))  # convert months to number-months
dataset1['Year_Month'] = dataset1['Year'] + '-' + dataset1['Month']  # for x-axis

# for standardizing dates (later)
# dataset1['Year_Month'] = pd.to_datetime(dataset1['Year_Month'])
```

```python
# Monthly Consumption of Fuel Over Time (2014-2023)
plt.figure(figsize=(20, 8))
plt.plot(dataset1['Year_Month'], dataset1['Consumption(million gallons)'], marker='o', linestyle='-', color='blue')
plt.title('Monthly Consumption of Fuel Over Time (2014-2023)')
plt.xlabel('Month')
plt.ylabel('Fuel Consumption (million gallons)')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
# Monthly Fuel Cost Over Time (2014-2023
plt.figure(figsize=(20, 8))
plt.plot(dataset1['Year_Month'], dataset1['Cost(million dollars)'], marker='o', linestyle='-', color='red')
plt.title('Monthly Fuel Cost Over Time (2014-2023)')
plt.xlabel('Month')
plt.ylabel('Fuel Cost (million dollars)')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
# Monthly Fuel Cost per Gallon Over Time (2014-2023)
plt.figure(figsize=(20, 8))
plt.plot(dataset1['Year_Month'], dataset1['Cost per Gallon (dollars)'], marker='o', linestyle='-', color='green')
plt.title('Monthly Fuel Cost per Gallon Over Time (2014-2023)')
plt.xlabel('Month')
plt.ylabel('Fuel Cost per Gallon (dollars)')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

## Dataset #2 National-Level Domestic Average Fare Series

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION 
dataset2 = pd.read_csv('Datasets/dataset2.csv')
dataset2['Year'] = dataset2['Year'].astype(str)  # convert years to strings
dataset2['Quarter'] = dataset2['Quarter'].astype(str)  # convert quarters to strings
dataset2 = dataset2[(dataset2['Year'] >= '2014') & (dataset2['Year'] <= '2023')]  # limit to ten years 2014-2023
dataset2['Year_Quarter'] = dataset2['Year'] + '-Q' + dataset2['Quarter']  # for x-axis
```

```python
# U.S. Average Ticket Price: Current vs. Inflation-Adjusted ($) Over Time (2014-2023)
plt.figure(figsize=(20, 8))
plt.plot(dataset2['Year_Quarter'], dataset2['U.S. Average (Current $) '], marker='o', linestyle='-', color='purple', label='U.S. Average (Current $)')
plt.plot(dataset2['Year_Quarter'], dataset2['U.S. Average (Inflation-Adjusted $) '], marker='o', linestyle='-', color='skyblue', label='U.S. Average (Inflation-Adjusted $)')
plt.title('U.S. Average Ticket Price: Current vs. Inflation-Adjusted ($) Over Time (2014-2023)')
plt.xlabel('Quarter')
plt.ylabel('Ticket Price (Dollars)')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()

plt.show()
```

## Dataset #3 Contiguous U.S. Precipitation

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION 
dataset3 = pd.read_csv('Datasets/dataset3.csv')
dataset3['Year'] = dataset3['Year'].astype(str)  # convert years to strings
dataset3['Month'] = dataset3['Month'].astype(str)  # convert months to strings
dataset3 = dataset3[(dataset3['Year'] >= '2014') & (dataset3['Year'] <= '2023')]  # limit to ten years 2014-2023
dataset3['Year_Month'] = dataset3['Year'] + '-' + dataset3['Month'].str.pad(2, fillchar='0')  # for x-axis

# for standardizing dates (later)
# dataset3['Year_Month'] = pd.to_datetime(dataset3['Year_Month'])
```

```python
# Average Precipitation Value Over Time (2014-2023)
plt.figure(figsize=(20, 8))
plt.plot(dataset3['Year_Month'], dataset3['Value'], marker='o', linestyle='-', color='darkorange', label='Value')
plt.title('Monthly Average U.S. Precipitation Over Time (2014-2023)')
plt.xlabel('Month')
plt.ylabel('Average Precipitation (inches)')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
# Precipitation Anomaly Over Time (2014-2023)
plt.figure(figsize=(20, 8))
plt.plot(dataset3['Year_Month'], dataset3['Anomaly'], marker='o', linestyle='-', color='teal', label='Anomaly')
plt.title('Precipitation Anomaly Over Time (2014-2023)')
plt.xlabel('Year-Month')
plt.ylabel('Anomaly')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()

plt.show()
```
